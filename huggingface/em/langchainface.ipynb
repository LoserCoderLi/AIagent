```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python

# 定义本地模型路径
local_model_path = "E:\\AAAAWork\\python\\models\\bce-embedding-base_v1"  # 替换为你的本地模型路径

# 初始化 HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(
    model_name=local_model_path,  # 使用本地模型路径
    model_kwargs={"device": "cpu"},  # 设置设备为CPU（或GPU）
    encode_kwargs={"normalize_embeddings": True},  # 可选参数，设置是否归一化嵌入
    cache_folder="/path/to/cache",  # 可选参数，设置缓存文件夹
    show_progress=True  # 显示进度条
)
```

```python
import os
from langchain_community.embeddings import DashScopeEmbeddings
# 从环境变量中获取API密钥，用于初始化 ChatTongyi 模型
api_key = os.getenv("KEY_TONGYI")
if not api_key:
    raise ValueError("API Key is not set. Please ensure that the 'KEY_TONGYI' environment variable is set.")

embedding=DashScopeEmbeddings(
            model="text-embedding-v1", dashscope_api_key=api_key
        )
```

```python
embedding
```

```python
# 读取文本文件内容
with open('./test.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# 将文本内容放入列表
texts = [text]
```

```python

# 生成文档的向量表示
document_embeddings = embeddings.embed_documents(texts)
```

```python

# 输出结果
print(document_embeddings)
```
